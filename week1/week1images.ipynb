# Plots for Week 1 Slides

## Import NymPy and MatPlotLib libraries

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
time = np.arange(0, 2*np.pi, 0.01);
```

```python
plt.plot(time, time)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.show()
```

```python
plt.plot(time, np.sin(time))
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.plot(time, np.sin(time*6))
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.plot(time, np.sin(time*12))
plt.xlabel('Time (Seconds)')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.plot(np.rad2deg(time), np.sin(time))
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.show()
```
